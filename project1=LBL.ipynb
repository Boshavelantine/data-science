```python
#reading data
import pandas as pd
d=pd.read_csv(r"C:\Users\Mohan\Downloads\CAR DETAILS FROM CAR DEKHO.csv")
d
```

```python
#checking null value
d.columns[d.isna().any()]
```

```python
#checking unique value in certain column
d["year"].unique()
```

```python
#giving value to x and y
x=d[["year"]]
y=d["selling_price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting data
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.7,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#fit
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#test
h=lr.predict(x_test)
h
```

```python
from sklearn.metrics import r2_score
```

```python
#comparing 
r2_score(h,y_test)
```
